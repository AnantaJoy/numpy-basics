```python
# what is an array in numpy?
# a numpy array is a grid of values, all of the same type, and is indexed by a tuple of nonnegative integers.
# the number of dimensions is the rank of the array; the shape of an array is a tuple of integers giving the size
# of the array along each dimension.
```

```python
# creating arrays
array_1d = np.array([1,2,3,4,4,5,5])
array_2d = np.array([[1,3,2,4,3,5,6,],
                     [7,6,5,4,3,2,3]])

array_3d = np.array([
    [[1, 2, 3, 4], [5, 6, 7, 8]],
    [[13, 14, 15, 16], [17, 18, 19, 20]]
])


# print the arrays
print(array_1d)
print(array_2d)
print(array_3d)

```

```python
# print the memory address of an array
print(id(array_1d))
# pointer to the memory location
print(array_1d.data)

# print the shape of an array
print(array_1d.shape)

# print the dimensions of an array
print(array_1d.ndim)

# print the size of an array
print(array_1d.size)

# print the data type of an array
print(array_1d.dtype)

# strides of an array
print(array_1d.strides)

print(array_2d.strides)

print(array_2d.flags)
```

```python
# converting data types in numpy
int_array = np.array([[1,2,3],[4,5,6],[4,2,9]], dtype=np.float64)

print(len(int_array))
int_array.astype(float)
```

```python
# creates an empty arrays
zeros = np.zeros(5)
ones = np.ones(10)
empty = np.empty(5)
ones_2d = np.ones((2,3))
full = np.full((3,4),7)
eye = np.eye(6)
random = np.random.random((2,4))

print(ones_2d)
print(random)
print(eye)
print(np.linspace(0,5,5))
```

```python
# basic array operations
x = np.array([[2,5,8],[3,7,4]])
y = np.array([2,5,8])
print(np.add(x,y))
print(np.subtract(x,y))
print(np.multiply(x,y))
print(np.divide(x,y))
```

```python
# statistical operations
print(x.sum())
print(x.min())
print(x.max())
print(x.max(axis=1))
print(x.cumsum())
print(x.mean())
print(np.median(x))
print(x.std())
```

```python
# concatenation of arrays

# create two 2D arrays
a = np.array([[1, 2], [3, 4]])
b = np.array([[5, 6], [7, 8]])

# concatenate the arrays along the first axis (rows)
c = np.concatenate((a, b), axis=0)

print(c)
```

```python
# vstack
# create two 2D arrays
a = np.array([[1, 2], [3, 4]])
b = np.array([[5, 6], [7, 8]])

# concatenate the arrays along the first axis (rows)
c = np.vstack((a, b))

print(c)
```

```python
# hstackimport numpy as np
# create two 2D arrays
a = np.array([[1, 2], [3, 4]])
b = np.array([[5, 6], [7, 8]])

# concatenate the arrays along the second axis (columns)
c = np.hstack((a, b))

print(c)
```

```python
# creating an image from array?
import numpy as np
import matplotlib.pyplot as plt

# Create a numpy array of the image data
image_data = np.array([[128, 8, 128],
                       [18, 187, 138],
                       [158, 38, 158]])

# Display the image
plt.imshow(image_data)
plt.show()

# Save the image to a file
plt.savefig('image.png')
```
